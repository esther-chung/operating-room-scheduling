```python
from gurobipy import *

# Create a new model
m = Model()

# Create binary variable for department i's usage of operating room j on day k
# x[i,j,k] = 1 if department i uses operating room j on day k, 0 otherwise
x = m.addVars(6, 5, 5, name="x", vtype = GRB.BINARY)
hours = m.addVars(6)
under_all = m.addVars(6)

target_ort = [92/190*213.5, 8/190*213.5, 48/190*213.5, 14/190*213.5, 10/190*213.5, 18/190*213.5]
time = [[9, 9, 9, 9, 7.5], [9, 9, 9, 9, 7.5], [9, 9, 9, 9, 7.5], [9, 9, 9, 9, 7.5], [9, 8, 8, 8, 6.5]]

for i in range(6):
    m.addConstr(hours[i] == sum(x[i, j, k] * time[j][k] for j in range(5) for k in range(5)))
    m.addConstr(under_all[i] >= target_ort[i] - hours[i])
    m.addConstr(under_all[i] >= 0.0)

floor = [0, 0, 1, 1, 2]
z = m.addVars(6,3,5, name="z", vtype=GRB.BINARY)
for i in range(6):
    for k in range(5):
        m.addConstr(sum(z[i,l,k] for l in range(3)) <= 1)

for i in range(6):
    for j in range(5):
        for k in range(5):
            m.addConstr(x[i,j,k] <= z[i,floor[j],k])

#Create objective function to minimize total under-allocation
m.setObjective(sum(under_all[i] / target_ort[i] for i in range(6)), GRB.MINIMIZE)

# Optimize the model
m.update()
m.optimize()

# Print the optimal value of the objective function
print("Optimal value of total under-allocation: ", m.objVal)
```
